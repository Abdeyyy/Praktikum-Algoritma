```python
tahun=int(input('Masukkan tahun :'))
if tahun % 4 ==0 and (tahun%100==0 and tahun%400!=0):
  print(str(tahun)+' bukan merupakan tahun kabisat')
else:
    if tahun%4==0:
      print(str(tahun)+' merupakan tahun kabisat')
    elif tahun%400==0:
      print(str(tahun)+' bukan merupakan tahun kabisat')
    else:
      print(str(tahun)+' bukan merupakan tahun kabisat')
```

```python
a1=int(input('Masukkan angka pertama : '))
a2=int(input('Masukkan angka kedua : '))
a3=int(input('Masukkan angka ketiga : '))

if a1<a2 :
  if a2<a3:
    print(f"{a1},{a2},{a3}"+' angka yang paling besar adalah', a3)
  else:
    print(f"{a1},{a2},{a3}"+' angka yang paling besar adalah', a2)
else:
    if a1<a3:
      print(f"{a1},{a2},{a3}"+' angka yang paling besar adalah', a3)
    else:
      print(f"{a1},{a2},{a3}"+' angka yang paling besar adalah', a1)
```

```python
k=input('Member/Non Member [Y/N] : ')
t=int(input('Total belanja : '))


if t>200000:
  if k=='Y':
    print('Mendapatkan diskon 6% sebesar = Rp', t*0.06)
  else:
    print('Mendapatkan diskon 4% sebesar = Rp', t*0.04)
elif t>100000:
  if k=='Y':
    print('Mendapatkan diskon 5% sebesar = Rp', t*0.05)
  else:
    print('Mendapatkan diskon 3% sebesar = Rp', t*0.03)
else:
  print('Tidak mendapatkan diskon')

```

```python
#weekly exercise 2
print('Selamat datang di Toko Maju Jaya')
print('Senin, 4 September 2023   '+'13.05')
print()
print('DAFTAR BELANJA:\n')
item1='Sabun'
item2="Shampo"
item3="Mie Instant"
item4='Detergen'
harga1=10000
harga2=15000
harga3=5000
harga4=17500
total= harga1+harga2+harga3+harga4
print(item1,'\t\t= Rp',harga1)
print(item2,'\t\t= Rp',harga2)
print(item3,'\t= Rp',harga3)
print(item4,"\t= Rp",harga4)
print()
print('Total\t= Rp',total)
print()
with open('Kuitansi.txt','w') as writefile:
  writefile.write('Selamat Datang di Toko Maju Jaya\n')
  writefile.write('\n')
  writefile.write('Kuitansi - 4 September 2023\n')
  writefile.write(item1+'\t\t= Rp '+str(harga1)+'\n')
  writefile.write(item2+'\t\t= Rp '+str(harga2)+'\n')
  writefile.write(item3+'\t= Rp '+str(harga3)+'\n')
  writefile.write(item4+"\t= Rp "+str(harga4))
  writefile.write('\n')
  writefile.write('Total Belanja \t= Rp '+str(total))
```

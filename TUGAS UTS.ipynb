```python
#TUGAS 1
NPM=str(input('Masukkan NPM Anda : '))    #memasukkan NPM mahasiswa
genap=0   #wadah/variabel untuk jumlah bilangan genap
ganjil=0  #wadah/variabel untuk jumlah bilangan ganjil

for i in NPM:   #indeks i untuk setiap bilangan di NPM
  if int(i)%2==0:   #apabila bilangan tersebut habis dibagi 2, maka masuk ke wadah genap
    genap+=1    #increment
  else:   #apabila bilangan tersebut tidak habis dibagi 2, maka masuk ke wadah ganjil
    ganjil+=1   #increment

print('Jumlah bilangan genap  : ', genap)   #mencetak jumlah bilangan genap
print('Jumlah bilangan ganjil : ', ganjil)  #mencetak jumlah bilangan ganjil
```

```python
class atk:
  def __init__(self):
    pass


#setter
  def info(self, nama, stok, hargaSatu):
    self.nama=nama
    self.stok=stok
    self.hargaSatu=hargaSatu
    self.hargaTotal=stok*hargaSatu


#getter
  def getItem(self):
    return{
        'Nama':self.nama,
        'Stok':self.stok,
        'Harga satuan':self.hargaSatu,
        'Harga total':self.hargaTotal
    }

bolpoint=atk()
bolpoint.info("Bolpoint",10,2000)

pensil=atk()
pensil.info('Pensil',10,1000)

penghapus=atk()
penghapus.info('Penghapus',10,500)

total=bolpoint.hargaTotal+pensil.hargaTotal+penghapus.hargaTotal
print(bolpoint.getItem())
print(pensil.getItem())
print(penghapus.getItem())
print(f'Total harga keseluruhan adalah : Rp{total}')
```

```python
#WEEKLY EXERCISE 1
t=int(input('Masukkan tinggi piramida : '))

for i in range(1, t+1):
  for j in range(t-i):
    print(' ', end="")
  for j in range(0, i):
    print('*', end=" ")
  print()
```

```python
#WEEKLY EXERCISE 2
n = int(input('Masukkan tinggi : '))

for i in range(n):
  for j in range(n-i-1):
    print(' ', end="")

  f = 1
  for j in range(i+1):
    print(f, end=" ")
    f = f * (i - j) // (j + 1)

  print()
```

```python
#WEEKLY EXERCISE 3
n = int(input('Masukkan baris segitiga Pascal: '))

C = 1
jumlah = 0
elemen = []

for j in range(n):
  elemen.append(C)
  jumlah += C
  C = C * (n - 1 - j) // (j + 1)

print(f"Baris: {elemen} jumlah angka dalam baris tersebut ada5lah: ", jumlah)
```
